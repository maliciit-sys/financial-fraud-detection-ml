```python
# =============================================================================
# NOTEBOOK 01: EDA ON NON-PCA PRIMARY DATASET (SPARKOV)
# =============================================================================
# Author: Muhammad Ali Tahir
# MS Data Science Program, Superior University Lahore
# Dataset: Sparkov Credit Card Transactions (Interpretable Features)
# =============================================================================


# %% [markdown]
# ## 1.1 Import Libraries

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
plt.style.use('seaborn-v0_8-whitegrid')

print("Libraries imported successfully!")

# %% [markdown]
# ## 1.2 Load Dataset

# %%
# Load training and test data
train_df = pd.read_csv('../../data/non_pca_primary/fraudTrain.csv')
test_df = pd.read_csv('../../data/non_pca_primary/fraudTest.csv')

print(f"Training set shape: {train_df.shape}")
print(f"Test set shape: {test_df.shape}")
print(f"Total transactions: {len(train_df) + len(test_df):,}")

# %% [markdown]
# ## 1.3 Dataset Overview

# %%
# Display first few rows
print("=" * 70)
print("FIRST 5 ROWS OF TRAINING DATA")
print("=" * 70)
print(train_df.head())

# %%
# Column information
print("=" * 70)
print("COLUMN INFORMATION")
print("=" * 70)
print(train_df.info())

# %%
# Statistical summary
print("=" * 70)
print("STATISTICAL SUMMARY (NUMERIC COLUMNS)")
print("=" * 70)
print(train_df.describe())

# %%
# Check for missing values
print("=" * 70)
print("MISSING VALUES")
print("=" * 70)
missing = train_df.isnull().sum()
missing_pct = (missing / len(train_df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Percentage': missing_pct})
print(missing_df[missing_df['Missing Count'] > 0] if missing.sum() > 0 else "No missing values!")

# %% [markdown]
# ## 1.4 Target Variable Analysis

# %%
# Fraud distribution
print("=" * 70)
print("FRAUD DISTRIBUTION")
print("=" * 70)
fraud_counts = train_df['is_fraud'].value_counts()
fraud_pct = train_df['is_fraud'].value_counts(normalize=True) * 100

print(f"Non-Fraud (0): {fraud_counts[0]:,} ({fraud_pct[0]:.2f}%)")
print(f"Fraud (1):     {fraud_counts[1]:,} ({fraud_pct[1]:.2f}%)")
print(f"\nClass Imbalance Ratio: 1:{fraud_counts[0]//fraud_counts[1]}")

# %%
# Visualize fraud distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot
colors = ['#2ecc71', '#e74c3c']
axes[0].bar(['Non-Fraud', 'Fraud'], fraud_counts.values, color=colors)
axes[0].set_title('Transaction Count by Class', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(fraud_counts.values):
    axes[0].text(i, v + 5000, f'{v:,}', ha='center', fontweight='bold')

# Pie chart
axes[1].pie(fraud_counts.values, labels=['Non-Fraud', 'Fraud'], autopct='%1.2f%%',
            colors=colors, explode=[0, 0.1], shadow=True)
axes[1].set_title('Fraud Percentage Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/fraud_distribution.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: fraud_distribution.png")

# %% [markdown]
# ## 1.5 Transaction Amount Analysis

# %%
# Amount statistics by fraud status
print("=" * 70)
print("TRANSACTION AMOUNT STATISTICS")
print("=" * 70)
amount_stats = train_df.groupby('is_fraud')['amt'].agg(['mean', 'median', 'std', 'min', 'max'])
amount_stats.index = ['Non-Fraud', 'Fraud']
print(amount_stats.round(2))

# %%
# Amount distribution visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Distribution of all amounts (log scale)
axes[0, 0].hist(train_df['amt'], bins=100, color='steelblue', edgecolor='black', alpha=0.7)
axes[0, 0].set_xlabel('Amount ($)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Distribution of Transaction Amounts', fontsize=12, fontweight='bold')
axes[0, 0].set_yscale('log')

# Box plot by fraud status
train_df.boxplot(column='amt', by='is_fraud', ax=axes[0, 1])
axes[0, 1].set_xlabel('Is Fraud')
axes[0, 1].set_ylabel('Amount ($)')
axes[0, 1].set_title('Amount Distribution by Fraud Status', fontsize=12, fontweight='bold')
plt.suptitle('')


# Distribution comparison (KDE)
train_df[train_df['is_fraud'] == 0]['amt'].plot(kind='kde', ax=axes[1, 0], label='Non-Fraud', color='#2ecc71')
train_df[train_df['is_fraud'] == 1]['amt'].plot(kind='kde', ax=axes[1, 0], label='Fraud', color='#e74c3c')
axes[1, 0].set_xlabel('Amount ($)')
axes[1, 0].set_ylabel('Density')
axes[1, 0].set_title('Amount Density by Class', fontsize=12, fontweight='bold')
axes[1, 0].legend()
axes[1, 0].set_xlim(0, 1000)

# Log-transformed amount distribution
axes[1, 1].hist(np.log1p(train_df[train_df['is_fraud'] == 0]['amt']), bins=50, alpha=0.7, label='Non-Fraud', color='#2ecc71')
axes[1, 1].hist(np.log1p(train_df[train_df['is_fraud'] == 1]['amt']), bins=50, alpha=0.7, label='Fraud', color='#e74c3c')
axes[1, 1].set_xlabel('Log(Amount + 1)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Log-Transformed Amount Distribution', fontsize=12, fontweight='bold')
axes[1, 1].legend()

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/amount_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: amount_analysis.png")


# %% [markdown]
# ## 1.6 Categorical Feature Analysis

# %%
# Category analysis
print("=" * 70)
print("TRANSACTION CATEGORIES")
print("=" * 70)
category_fraud = train_df.groupby('category').agg({
    'is_fraud': ['sum', 'count', 'mean']
}).round(4)
category_fraud.columns = ['Fraud_Count', 'Total_Count', 'Fraud_Rate']
category_fraud = category_fraud.sort_values('Fraud_Rate', ascending=False)
print(category_fraud)

# %%
# Category visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Transaction count by category
category_counts = train_df['category'].value_counts()
axes[0].barh(category_counts.index, category_counts.values, color='steelblue')
axes[0].set_xlabel('Transaction Count')
axes[0].set_title('Transactions by Category', fontsize=12, fontweight='bold')
axes[0].invert_yaxis()

# Fraud rate by category
fraud_rate_by_cat = train_df.groupby('category')['is_fraud'].mean().sort_values(ascending=True)
colors = plt.cm.RdYlGn_r(np.linspace(0.2, 0.8, len(fraud_rate_by_cat)))
axes[1].barh(fraud_rate_by_cat.index, fraud_rate_by_cat.values * 100, color=colors)
axes[1].set_xlabel('Fraud Rate (%)')
axes[1].set_title('Fraud Rate by Category', fontsize=12, fontweight='bold')
axes[1].axvline(x=train_df['is_fraud'].mean() * 100, color='red', linestyle='--', label='Overall Avg')
axes[1].legend()

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/category_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: category_analysis.png")

# %% [markdown]
# ## 1.7 Gender Analysis

# %%
# Gender distribution and fraud rate
print("=" * 70)
print("GENDER ANALYSIS")
print("=" * 70)
gender_analysis = train_df.groupby('gender').agg({
    'is_fraud': ['sum', 'count', 'mean'],
    'amt': 'mean'
}).round(4)
gender_analysis.columns = ['Fraud_Count', 'Total_Count', 'Fraud_Rate', 'Avg_Amount']
print(gender_analysis)

# %%
# Gender visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Count by gender
gender_counts = train_df['gender'].value_counts()
axes[0].pie(gender_counts.values, labels=['Female', 'Male'], autopct='%1.1f%%',
            colors=['#ff9999', '#66b3ff'], explode=[0.02, 0.02])
axes[0].set_title('Transaction Distribution by Gender', fontsize=12, fontweight='bold')

# Fraud rate by gender
fraud_by_gender = train_df.groupby('gender')['is_fraud'].mean() * 100
axes[1].bar(['Female', 'Male'], fraud_by_gender.values, color=['#ff9999', '#66b3ff'])
axes[1].set_ylabel('Fraud Rate (%)')
axes[1].set_title('Fraud Rate by Gender', fontsize=12, fontweight='bold')
for i, v in enumerate(fraud_by_gender.values):
    axes[1].text(i, v + 0.02, f'{v:.2f}%', ha='center', fontweight='bold')

# Average amount by gender and fraud status
gender_fraud_amt = train_df.groupby(['gender', 'is_fraud'])['amt'].mean().unstack()
gender_fraud_amt.plot(kind='bar', ax=axes[2], color=['#2ecc71', '#e74c3c'])
axes[2].set_xlabel('Gender')
axes[2].set_ylabel('Average Amount ($)')
axes[2].set_title('Average Amount by Gender & Fraud Status', fontsize=12, fontweight='bold')
axes[2].legend(['Non-Fraud', 'Fraud'])
axes[2].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/gender_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: gender_analysis.png")

# %% [markdown]
# ## 1.8 Temporal Analysis

# %%
# Parse datetime
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])
train_df['hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['day_of_week'] = train_df['trans_date_trans_time'].dt.dayofweek
train_df['month'] = train_df['trans_date_trans_time'].dt.month

# %%
# Hourly analysis
print("=" * 70)
print("HOURLY FRAUD ANALYSIS")
print("=" * 70)
hourly_fraud = train_df.groupby('hour')['is_fraud'].agg(['sum', 'count', 'mean'])
hourly_fraud.columns = ['Fraud_Count', 'Total_Count', 'Fraud_Rate']
print(hourly_fraud.round(4))

# %%
# Temporal visualization
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Hourly transaction volume
hourly_counts = train_df.groupby('hour').size()
hourly_fraud_rate = train_df.groupby('hour')['is_fraud'].mean() * 100

ax1 = axes[0, 0]
ax2 = ax1.twinx()
ax1.bar(hourly_counts.index, hourly_counts.values, alpha=0.7, color='steelblue', label='Transactions')
ax2.plot(hourly_fraud_rate.index, hourly_fraud_rate.values, color='red', marker='o', linewidth=2, label='Fraud Rate')
ax1.set_xlabel('Hour of Day')
ax1.set_ylabel('Transaction Count', color='steelblue')
ax2.set_ylabel('Fraud Rate (%)', color='red')
ax1.set_title('Hourly Transaction Volume & Fraud Rate', fontsize=12, fontweight='bold')

# Day of week analysis
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
daily_fraud_rate = train_df.groupby('day_of_week')['is_fraud'].mean() * 100
axes[0, 1].bar(day_names, daily_fraud_rate.values, color='coral')
axes[0, 1].set_xlabel('Day of Week')
axes[0, 1].set_ylabel('Fraud Rate (%)')
axes[0, 1].set_title('Fraud Rate by Day of Week', fontsize=12, fontweight='bold')

# Monthly analysis
monthly_fraud = train_df.groupby('month')['is_fraud'].mean() * 100
axes[1, 0].plot(monthly_fraud.index, monthly_fraud.values, marker='o', linewidth=2, color='purple')
axes[1, 0].set_xlabel('Month')
axes[1, 0].set_ylabel('Fraud Rate (%)')
axes[1, 0].set_title('Monthly Fraud Rate Trend', fontsize=12, fontweight='bold')
axes[1, 0].set_xticks(range(1, 13))

# Heatmap: Hour vs Day
pivot_fraud = train_df.pivot_table(values='is_fraud', index='day_of_week', columns='hour', aggfunc='mean')
sns.heatmap(pivot_fraud * 100, ax=axes[1, 1], cmap='YlOrRd', annot=False, fmt='.2f',
            yticklabels=day_names, cbar_kws={'label': 'Fraud Rate (%)'})
axes[1, 1].set_title('Fraud Rate: Hour vs Day of Week', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Hour of Day')
axes[1, 1].set_ylabel('Day of Week')

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/temporal_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: temporal_analysis.png")

# %% [markdown]
# ## 1.9 Geographic Analysis

# %%
# State analysis
print("=" * 70)
print("TOP 10 STATES BY FRAUD RATE")
print("=" * 70)
state_fraud = train_df.groupby('state').agg({
    'is_fraud': ['sum', 'count', 'mean']
})
state_fraud.columns = ['Fraud_Count', 'Total_Count', 'Fraud_Rate']
state_fraud = state_fraud[state_fraud['Total_Count'] >= 1000]  # Filter low volume states
state_fraud = state_fraud.sort_values('Fraud_Rate', ascending=False)
print(state_fraud.head(10).round(4))

# %%
# Geographic visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Top 15 states by transaction volume
top_states = train_df['state'].value_counts().head(15)
axes[0].barh(top_states.index, top_states.values, color='steelblue')
axes[0].set_xlabel('Transaction Count')
axes[0].set_title('Top 15 States by Transaction Volume', fontsize=12, fontweight='bold')
axes[0].invert_yaxis()

# Top 15 states by fraud rate (min 1000 transactions)
top_fraud_states = state_fraud.head(15)['Fraud_Rate'] * 100
axes[1].barh(top_fraud_states.index, top_fraud_states.values, color='coral')
axes[1].set_xlabel('Fraud Rate (%)')
axes[1].set_title('Top 15 States by Fraud Rate (min 1000 txns)', fontsize=12, fontweight='bold')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/geographic_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: geographic_analysis.png")

# %% [markdown]
# ## 1.10 Age Analysis

# %%
# Calculate age from DOB
train_df['dob'] = pd.to_datetime(train_df['dob'])
train_df['age'] = (train_df['trans_date_trans_time'] - train_df['dob']).dt.days // 365

# %%
# Age statistics
print("=" * 70)
print("AGE STATISTICS BY FRAUD STATUS")
print("=" * 70)
age_stats = train_df.groupby('is_fraud')['age'].agg(['mean', 'median', 'std', 'min', 'max'])
age_stats.index = ['Non-Fraud', 'Fraud']
print(age_stats.round(2))

# %%
# Age visualization
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Age distribution
axes[0].hist(train_df[train_df['is_fraud'] == 0]['age'], bins=50, alpha=0.7, label='Non-Fraud', color='#2ecc71')
axes[0].hist(train_df[train_df['is_fraud'] == 1]['age'], bins=50, alpha=0.7, label='Fraud', color='#e74c3c')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Age Distribution by Fraud Status', fontsize=12, fontweight='bold')
axes[0].legend()

# Age groups fraud rate
train_df['age_group'] = pd.cut(train_df['age'], bins=[0, 25, 35, 45, 55, 65, 100],
                               labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
age_group_fraud = train_df.groupby('age_group')['is_fraud'].mean() * 100
axes[1].bar(age_group_fraud.index.astype(str), age_group_fraud.values, color='steelblue')
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('Fraud Rate (%)')
axes[1].set_title('Fraud Rate by Age Group', fontsize=12, fontweight='bold')

# Box plot
train_df.boxplot(column='age', by='is_fraud', ax=axes[2])
axes[2].set_xlabel('Is Fraud')
axes[2].set_ylabel('Age')
axes[2].set_title('Age Distribution by Fraud Status', fontsize=12, fontweight='bold')
plt.suptitle('')

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/age_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: age_analysis.png")

# %% [markdown]
# ## 1.11 Merchant Analysis

# %%
# Top merchants by fraud
print("=" * 70)
print("TOP 10 MERCHANTS BY FRAUD COUNT")
print("=" * 70)
merchant_fraud = train_df.groupby('merchant').agg({
    'is_fraud': ['sum', 'count', 'mean']
})
merchant_fraud.columns = ['Fraud_Count', 'Total_Count', 'Fraud_Rate']
merchant_fraud = merchant_fraud.sort_values('Fraud_Count', ascending=False)
print(merchant_fraud.head(10).round(4))

# %% [markdown]
# ## 1.12 Distance Analysis (Customer to Merchant)

# %%
# Calculate distance between customer and merchant
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(lat1, lon1, lat2, lon2):
    """Calculate the distance between two points on Earth using Haversine formula"""
    R = 6371  # Earth's radius in kilometers

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))

    return R * c

# Calculate distance
train_df['distance_km'] = haversine_distance(
    train_df['lat'], train_df['long'],
    train_df['merch_lat'], train_df['merch_long']
)

# %%
# Distance statistics
print("=" * 70)
print("DISTANCE STATISTICS (Customer to Merchant)")
print("=" * 70)
distance_stats = train_df.groupby('is_fraud')['distance_km'].agg(['mean', 'median', 'std', 'min', 'max'])
distance_stats.index = ['Non-Fraud', 'Fraud']
print(distance_stats.round(2))

# %%
# Distance visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distance distribution (log scale)
axes[0].hist(np.log1p(train_df[train_df['is_fraud'] == 0]['distance_km']), bins=50, alpha=0.7, label='Non-Fraud', color='#2ecc71')
axes[0].hist(np.log1p(train_df[train_df['is_fraud'] == 1]['distance_km']), bins=50, alpha=0.7, label='Fraud', color='#e74c3c')
axes[0].set_xlabel('Log(Distance + 1) km')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Distance Distribution by Fraud Status', fontsize=12, fontweight='bold')
axes[0].legend()

# Box plot
train_df.boxplot(column='distance_km', by='is_fraud', ax=axes[1])
axes[1].set_xlabel('Is Fraud')
axes[1].set_ylabel('Distance (km)')
axes[1].set_title('Distance Distribution by Fraud Status', fontsize=12, fontweight='bold')
plt.suptitle('')

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/distance_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: distance_analysis.png")

# %% [markdown]
# ## 1.13 Correlation Analysis

# %%
# Select numeric columns for correlation
numeric_cols = ['amt', 'lat', 'long', 'city_pop', 'unix_time', 'merch_lat', 'merch_long',
                'hour', 'day_of_week', 'month', 'age', 'distance_km', 'is_fraud']
corr_matrix = train_df[numeric_cols].corr()

# %%
# Correlation with target
print("=" * 70)
print("CORRELATION WITH TARGET (is_fraud)")
print("=" * 70)
target_corr = corr_matrix['is_fraud'].drop('is_fraud').sort_values(key=abs, ascending=False)
print(target_corr.round(4))

# %%
# Correlation heatmap
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Full correlation matrix
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='RdBu_r', center=0, ax=axes[0],
            square=True, linewidths=0.5)
axes[0].set_title('Correlation Matrix (Numeric Features)', fontsize=12, fontweight='bold')

# Correlation with target
colors = ['#e74c3c' if x < 0 else '#2ecc71' for x in target_corr.values]
axes[1].barh(target_corr.index, target_corr.values, color=colors)
axes[1].set_xlabel('Correlation with is_fraud')
axes[1].set_title('Feature Correlation with Target', fontsize=12, fontweight='bold')
axes[1].axvline(x=0, color='black', linestyle='-', linewidth=0.5)

plt.tight_layout()
plt.savefig('../../outputs/non_pca_primary/eda/correlation_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Saved: correlation_analysis.png")

# %% [markdown]
# ## 1.14 Summary Statistics

# %%
print("=" * 70)
print("EDA SUMMARY")
print("=" * 70)
print(f"""
DATASET OVERVIEW:
- Total Training Transactions: {len(train_df):,}
- Total Test Transactions: {len(test_df):,}
- Features: {len(train_df.columns)} columns
- Date Range: {train_df['trans_date_trans_time'].min()} to {train_df['trans_date_trans_time'].max()}

CLASS DISTRIBUTION:
- Non-Fraud: {fraud_counts[0]:,} ({fraud_pct[0]:.2f}%)
- Fraud: {fraud_counts[1]:,} ({fraud_pct[1]:.2f}%)
- Imbalance Ratio: 1:{fraud_counts[0]//fraud_counts[1]}

KEY FINDINGS:
1. Transaction Amount:
   - Fraud transactions have higher average amount (${train_df[train_df['is_fraud']==1]['amt'].mean():.2f})
   - vs Non-fraud (${train_df[train_df['is_fraud']==0]['amt'].mean():.2f})

2. Temporal Patterns:
   - Highest fraud rate hours: Late night / Early morning
   - Weekend vs Weekday patterns exist

3. Categories:
   - Highest fraud rate: {category_fraud.index[0]} ({category_fraud.iloc[0]['Fraud_Rate']*100:.2f}%)
   - Lowest fraud rate: {category_fraud.index[-1]} ({category_fraud.iloc[-1]['Fraud_Rate']*100:.2f}%)

4. Demographics:
   - Age range: {train_df['age'].min()} to {train_df['age'].max()} years

5. Geographic:
   - Distance to merchant varies between fraud and non-fraud transactions
""")

# %% [markdown]
# ## 1.15 Save Processed Data Info

# %%
# Save feature info for model training
feature_info = {
    'numeric_features': ['amt', 'city_pop', 'age', 'distance_km', 'hour', 'day_of_week', 'month'],
    'categorical_features': ['category', 'gender', 'state'],
    'target': 'is_fraud',
    'drop_columns': ['Unnamed: 0', 'trans_date_trans_time', 'cc_num', 'merchant', 'first', 'last',
                     'street', 'city', 'zip', 'lat', 'long', 'job', 'dob', 'trans_num',
                     'unix_time', 'merch_lat', 'merch_long'],
    'train_samples': len(train_df),
    'test_samples': len(test_df),
    'fraud_rate': train_df['is_fraud'].mean()
}

import json
with open('../../outputs/non_pca_primary/eda/feature_info.json', 'w') as f:
    json.dump(feature_info, f, indent=2)

print(" Saved: feature_info.json")
print("\nEDA Complete! Ready for model training.")
```
